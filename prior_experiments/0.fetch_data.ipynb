```python
import numpy as np
import pandas as pd

from clearinghouse_util import fetch_all_case_data
```

```python
df = pd.read_csv("clearinghouse.csv")
df.head()
```

```python
%%time

results = fetch_all_case_data(df)
results
```

```python
df["response"] = results
```

```python
df['response_head'] = df['response'].apply(lambda d: next(iter(d)) if isinstance(d, dict) and d else None)
```

```python
df['response_head'].value_counts()
```

```python
retries = df[df["response_head"] == "error"].reset_index()
len(retries)
```

```python
%%time

retry_results = fetch_all_case_data(retries)
```

```python
retries["response"] = retry_results
retries['response_head'] = retries['response'].apply(lambda d: next(iter(d)) if isinstance(d, dict) and d else None)
retries['response_head'].value_counts()
```

```python
retries.set_index('index', inplace=True)
retries.head()
```

```python
df.update(retries)
```
