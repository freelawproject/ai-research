# Assign SCALES Label

To eveluate the SCALES model, we need to first assign labels to the eval dataset based on the SCALES ontology. I used GPT and SCALES model to generate the predictions first, and then used the predictions to help speed-up the label creating process.

# Import libary

```python
import numpy as np
import pandas as pd

from gpt_util import predict
```

# Use GPT as first pass to assign labels to entry

```python
df = pd.read_csv("data/entry.csv")
df.head()
```

```python
%%time

predictions = predict(df)
```

```python
predictions
```

```python
prediction_df = pd.DataFrame(predictions)
prediction_df["label"] = prediction_df["label"].str.replace(r'["\']', '', regex=True)
prediction_df
```

```python
assert len(df) == len(prediction_df)
```

```python
output_df = df.merge(prediction_df, how="left", on="unique_id")
output_df.head()
```

```python
output_df["input_tokens"].sum()
```

```python
output_df["output_tokens"].sum()
```

```python
output_df = output_df.rename(columns={"label":"gpt_pred"})
output_df.head()
```

# Use SCALES model to make predictions

```python
from transformers import pipeline

!pip install -U datasets -q
from datasets import Dataset
```

```python
pipe = pipeline("text-classification", model="scales-okn/docket-classification")
```

```python
%%time
dataset = Dataset.from_pandas(output_df[['text', 'unique_id']])
predictions = pipe(dataset["text"])
```

```python
predictions[:5]
```

```python
assert len(predictions) == len(output_df)
```

```python
predictions_df = pd.DataFrame(predictions)
predictions_df.head()
```

```python
predictions_df["unique_id"] = output_df["unique_id"]
predictions_df.head()
```

```python
final_df = output_df.merge(predictions_df, how="left", on="unique_id")
final_df.head()
```

```python
final_df = final_df.rename(columns={"label": "scales_pred",
                                    "score": "scales_score"})
final_df.head()
```

# Save the data for manual review and manually assign labels

```python
final_df.to_csv("data/entry_label_review.csv", index=False)
```
