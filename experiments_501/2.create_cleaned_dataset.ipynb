```python
df['note'] = df['filename'].map(notes)
```

```python
notes_df = pd.DataFrame([
    {
        "note": v,
        "citing_cluster_id": int(k.split(".")[1].split("_")[0]),
        "cited_cluster_id": int(k.split("_")[-1].replace(".txt", ""))
    }
    for k, v in notes.items()
])

notes_df.head()
```

```python
df = df.merge(
    notes_df,
    on=['citing_cluster_id', 'cited_cluster_id'],
    how='left',
    suffixes=('', '_note')
)
```

```python
df['note'] = df['note'].combine_first(df['note_note'])
df.drop(columns='note_note', inplace=True)
```

```python
df['use_full_opinion'] = df['note'].notna().astype(int)
df.head()
```

```python
df['use_full_opinion'].value_counts()
```

```python
result = df.merge(citing[["citing_cluster_id", "citing_url", "citing_opinions", "citing_filenames"]], how="left", on="citing_cluster_id")
result.head()
```
