# Check for any empty excerpts

```python
import os

# Set the folder path
folder_path = "data/excerpts/"

# List all txt files in the folder
txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

# Check each file for being blank
blank_files = []
for file in txt_files:
    file_path = os.path.join(folder_path, file)
    if os.path.getsize(file_path) == 0:  # Checks if file size is 0 bytes
        blank_files.append(file)

# Print the result
if blank_files:
    print("Blank txt files found:", blank_files)
else:
    print("No blank txt files found.")
```
