# Baseline

To establish a baseline, I selected GPT-4o-mini as the model to establish a baseline for model performance. Using the model predictions as guidance, I also created the sup class label for the dataset manually

# Import libraries

```python
import numpy as np
import pandas as pd

from gpt_util import predict
```

# Load the data

```python
df = pd.read_csv("data/entry_label_assign.csv")
df = df[["unique_id", "text"]]
df.head()
```

```python
len(df)
```

# Predict using 4o with a generic prompt

```python
%%time

predictions = predict(df)
```

```python
predictions
```

```python
prediction_df = pd.DataFrame(predictions)
prediction_df["label"] = prediction_df["label"].str.replace(r'["\']', '', regex=True)
prediction_df.head()
```

```python
assert len(df) == len(prediction_df)
```

```python
output_df = df.merge(prediction_df, how="left", on="unique_id")
output_df.head()
```

```python
output_df["input_tokens"].sum()
```

```python
output_df["output_tokens"].sum()
```

```python
output_df = output_df.rename(columns={"label":"gpt_pred_sup"})
output_df.head()
```

# Save the predictions for manually adding the labels

```python
output_df.to_csv("data/entry_label_assign_gpt.csv", index=False)
```

# Look at stats for the "Other" class by using GPT and a generic prompt to assign sub class labels to "Other" sup class documents

```python
df = pd.read_csv("data/entry_label_assign_gpt_labelled.csv")
len(df)
```

```python
df = df[df["flp_sup"] == "Other"]
len(df)
```

```python
%%time

predictions = predict(df)
```

```python
predictions
```

```python
prediction_df = pd.DataFrame(predictions)
prediction_df["label"] = prediction_df["label"].str.replace(r'["\']', '', regex=True)
prediction_df.head()
```

```python
assert len(df) == len(prediction_df)
```

```python
output_df = df.merge(prediction_df, how="left", on="unique_id")
output_df.head()
```

```python
output_df = output_df.rename(columns={"label":"gpt_pred_other"})
output_df.head()
```

```python
output_df["gpt_pred_other"].value_counts()
```
